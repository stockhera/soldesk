```python
#네이버증권에서 달러환율 데이터 크롤링하기
import requests # urllib.request as req 같은 것
from bs4 import BeautifulSoup

import pandas #데이터분석

import matplotlib as mpl #시각화
import matplotlib.pyplot as plt #시각화

#데이터 긁어오는 스크래핑
url = "https://www.weather.go.kr/w/observation/land/city-obs.do"
source = requests.get(url)

#데이터를 html로 변환하는 크롤링
soup = BeautifulSoup(source.content, 'html.parser')
table =soup.find("table",{"class" : "table-col"})
#print(table)
data = []
for tr in table.find_all('tr'):
    ths = list(tr.find_all('th'))
    tds = list(tr.find_all('td'))
    #print(tds)
    for th in ths:
        if th.find('a'):
            point = th.find('a').string
            temp = tds[4].string
            humidity = tds[9].string
            data.append([point, temp, humidity])
#csv: comma seperate value / tsv:tab seperate value
with open("weather.csv", "w", encoding="utf-8") as f:
    f.write('지역, 온도,습도\n')
    for l in data:
        f.write(f"{l[0]},{l[1]},{l[2]}\n")
df = pandas.read_csv("weather.csv", encoding="utf-8", index_col='지역')
#print(df)
city_df = df.loc[['서울','인천','대전','대구','광주','부산','울산']]
#print(city_df)
font_name = mpl.font_manager.FontProperties(fname="C:\Windows\Fonts\malgun.ttf").get_name()
mpl.rc('font', family=font_name)
ax = city_df.plot(kind='bar', title='weather', figsize=(12,4))

plt.show()

## 한글지원하는 syntax github에서 확인하기
```

```python
#유니버셜 함수 (UFuncs)
l1 = [1,2,3]
l2 = [3,2,1]
x = np.array([1,2,3])
y = np.array([3,2,1])
print(l1 + l2 ) #확장
print(x + y) #덧셈연산
print(np.add(x,y))
# + 연산자 : np.add() 유니버셜 함수
print(np.add(x,y))
```

```python
def add(l1, l2):
    l4 = []
    for idx in range (0, len(l1)):
        l4.append(l1[idx] + l2[idx])
    return l4
```

```python
result = add (l1, l2)
print(result)
```
